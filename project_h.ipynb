```python
#fill all null value to 0
df = df.fillna(0)
```

```python
# replace $ and str to floate 
# I did it two time that's why this error is showing
df["Total Charged"] = df["Total Charged"].str.replace('$','').astype(float)
```

```python
df.head()
```

```python
#calculte the money spend on amazon
df['Total Charged'].sum()
```

```python
#Avarage purchase on amazon
df['Total Charged'].mean()
```

```python
# finding median
df['Total Charged'].median()
```

```python
# high purchase on amazon
df['Total Charged'].max()
```

```python
# Minimum purchase
df['Total Charged'].min()
```

```python
# now we wants to do calculation for 'Tax charged' to now we should clean it than after we can do math on 'Tax Charged'
# cleaning 
df["Tax Charged"] = df["Tax Charged"].str.replace('$','').astype(float)
df.head()
```

```python
# find the total tax 
df['Tax Charged'].sum()
```

```python
# finding the total per iteam tax 
df['Tax Charged'].sum()/df['Total Charged'].sum()
```

```python
# we have to set 'order date' type because there is order date is behaving as a string so we have to convert  
# so that computer read as a date 
df['Order Date']= pd.to_datetime(df['Order Date'])
```

```python
#checking
df.head()
```

```python
# ploting bar graph to better visualling and analysis
df.plot.bar(x='Order Date', y='Total Charged')
```

```python
# In the above figure there we having some diffcultes to view graph so that we set figure size
df.plot.bar(x='Order Date', y='Total Charged', figsize=(15,10))
```

```python
#So we have to analysis more efficent way. so we are grouping 'Order date' and sum 'Totel Charged'
daily_orders = df.groupby('Order Date').sum()["Total Charged"]
daily_orders.head()
```

```python
# ploting bar graph od daily_order so we can se efficly visulise
daily_orders.plot.bar(x='Order Date', y='Total Charged', figsize=(15,8))
```
